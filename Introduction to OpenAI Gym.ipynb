The **gym.make(env)** function call within the Gym library is a critical step in creating an environment for reinforcement learning tasks. Here's a breakdown of what it does:

1. ****gym.make()****: This function is a part of the Gym library and is used to instantiate an environment based on the specified environment ID (**env**). It's the primary method for creating environments within Gym.

2. **Environment ID (**env**)**: The **env** parameter represents the identifier for the specific environment you want to create. It could be a string denoting the name of a pre-defined environment provided by Gym, such as 'CartPole-v1', 'MountainCar-v0', 'Pong-v0', etc. Each environment simulates a particular task or scenario for reinforcement learning.

   For example:
   - **'CartPole-v1'** simulates the cart-pole balancing task.
   - **'MountainCar-v0'** simulates a car struggling to reach a flag on top of a mountain.
   - **'Pong-v0'** simulates the classic game of Pong.

3. **Creation of the Environment**: When **gym.make(env)** is called, Gym will create an instance of the specified environment, initializing it to its starting state.

4. **Interface Consistency**: Gym provides a consistent interface across different environments. Once an environment is created using **gym.make()**, it allows interaction through standard methods like **reset()**, **step()**, and accessing properties like **observation_space** and **action_space**.

   - **reset()**: Resets the environment to its initial state and returns the initial observation.
   - **step(action)**: Takes an action in the environment and returns the next observation, reward, a boolean indicating whether the episode is done, and additional information.
   - **observation_space**: Describes the space of possible observations the agent can receive.
   - **action_space**: Defines the set of possible actions the agent can take.


```python
# Initalize/Create an environment
env_id = "CartPole-v1"
env = gym.make(env_id)

# Reset the env to it's initial state
env.reset()

# Set an env seed for reproducibility
env.seed(42)
```

As mentioned before, each environment is identified by an unique id, which is represented by environment's name and version number.

```python
env_methods = [m for m in dir(env) if not m.startswith("_")]
for index, method in enumerate(env_methods):
    print(f"{index+1:2}. {method}")
```

Probably one of the most important methods available within the environment is the **step()** method, because it is the **bridge between actions, rewards and observations.** It allows you to execute an action in the environment.

```python
action = 0                  # For now let's have a dumb agent

# Game configs
total_reward = 0.0
total_steps = 0

done = False
while not done:

    obs, reward, done, info = env.step(action)

    total_reward += reward
    total_steps += 1

print(f"Episode completed in {total_steps} steps.\nTotal rewards accumulated: {total_reward}")
```

$Step$ method: This method returns a tuple of four quantities. The **observation, the reward, done and info.**

1. **Observation**: The new observation of the agent.

2. **Reward**: The reward gained by the agent.

3. **Done**: A boolean if the game reaches its end.

4. **Info**: Extra information about the environment.


```python
obs, reward, done, info = env.step(action)

print(f"Observations   : {obs}")
print(f"Reward Gained  : {reward}")
print(f"Game Over?     : {done}")
print(f"Additional Info: {info}")
```

# **Action & Observation Space 🕹️**

**Action Space**: This refers to the complete set of actions an agent can perform within an environment. This space can manifest as either continuous, where actions are represented by real numbers within a range, or discrete, where actions are represented by distinct and separate values.

**Observation Space**: This represents the information furnished by the environment to the agent at any given time step. Within the realm of OpenAI Gym, these observation spaces are categorized into three primary types:

* **Box**: This category encapsulates the visual observation space, typically represented as an image capturing the environment's visual state. In essence, it provides an image-based perspective of the environment, akin to what a human would perceive.

* **Discrete**: Here, the observation space presents discrete and distinct information supplied by the environment, often in the form of numerical values or categorical data. This form of observation space furnishes specific, separate pieces of information relevant to the environment's state.

* **Tuple**: This category encompasses a composite observation space, combining multiple individual spaces into a cohesive unit. It amalgamates diverse types of information, potentially including visual data (like images), numerical values, categorical information, or any other relevant data types essential for understanding the environment's state.

image.png

In OpenAI Gym's observation spaces, regardless of their types—be it **'Box', 'Discrete', or 'Tuple'**—two fundamental methods are consistently present within the Space class:

1. **contains()**: This method verifies whether a specific observation is part of the observation space or not. It acts as a check to ascertain if a particular observation is within the defined space, ensuring the validity of observations generated or encountered during agent interaction within the environment.

2. **sample()**: The `sample()` method randomly retrieves an observation from the observation space. It effectively returns any valid observation within the space, allowing for random sampling of the space's content. This function is particularly useful during the exploration phase of reinforcement learning when agents seek to gather diverse experiences by randomly selecting observations from the space (the same for goes for action space).

These core methods, inherent to the Space class across all observation space types, provide essential functionality for validating and interacting with observations within Gym environments. The `contains()` method ensures the conformity of observations, while `sample()` facilitates random extraction of observations for exploration and learning purposes, contributing significantly to the adaptability and robustness of agents interacting with diverse environments.

# **Wrappers** 🎁

**Gym's wrappers** are **powerful tools** allowing users to **augment the functionality** of environments by encapsulating and extending their logic.

These wrappers offer the **flexibility to modify or enhance an environment's behavior**, enabling diverse functionalities such as:

1. **Image Processing**: By applying wrappers, you can preprocess visual observations, implementing image transformations, filters, or feature extraction methods before these observations are passed to the agent. This preprocessing can significantly impact an agent's ability to comprehend and learn from visual data.

2. **Observation Storage**: Wrappers can be used to store observations in a relay buffer or any other form of memory. This can aid in creating replay buffers for experience replay techniques in reinforcement learning, facilitating improved learning and sample efficiency.

3. **Feature Extraction and State Representation**: They enable the extraction of specific features or the creation of custom state representations from observations. This allows for the adaptation of observations into a format that better suits an agent's learning process.

4. **Reward Manipulation**: Wrappers can modify the reward structure or introduce custom reward functions tailored to specific learning objectives or experimental requirements.

5. **Control and Monitoring**: They provide means to control the environment's behavior or introduce monitoring mechanisms to track specific aspects of an agent's interaction with the environment.

6. **Temporal Frame Stacking**: Wrappers can stack multiple frames together to create temporal sequences of observations, providing the agent with a sense of motion or temporal continuity.

image.png

The **Wrapper** class in Gym serves as an extension of the environment class, offering three distinct child classes: the **ObservationWrapper**, **RewardWrapper**, and **ActionWrapper**.

Each of these child classes offers a specialized functionality:

1. **Observation Wrapper**: This wrapper allows users to redefine or modify the observation function within an environment. By subclassing the **ObservationWrapper** and overriding specific methods, users can preprocess, transform, or manipulate observations before they are presented to the agent, tailoring the observation data to better suit the learning process.

2. **Reward Wrapper**: Similarly, the **RewardWrapper** enables the redefinition of the reward function. Overriding relevant methods in this wrapper permits customization or modification of the reward structure within the environment. Users can introduce custom reward schemes or manipulate reward values to align with specific learning objectives or experimental requirements.

3. **Action Wrapper**: The **ActionWrapper** empowers users to redefine the action function within an environment. By overriding designated methods in this wrapper, users can modify or transform the action space, potentially implementing constraints, preprocessing actions, or altering the action selection process before they are executed within the environment.

---
Let's see an example,

```python
class RandomActionWrapper(ActionWrapper):

    def __init__(self, env, epsilon = 0.1):
        super(RandomActionWrapper, self).__init__(env)

        self.env = env
        self.eps = epsilon

    def action(self, action):

        if random.random() < self.eps:
            return self.env.action_space.sample()

        return action
```

```python
# Initilize the env
env = gym.make(env_id)

# Apply the Wrapper
env = RandomActionWrapper(env, epsilon = 0.5)    # Try different values

# Reset env to initial state
env.reset()

# Game configs
total_reward = 0.0
total_steps = 0

done = False
while not done:
    _, reward, done, _ = env.step(0)            # A dumb agent

    total_reward += reward
    total_steps += 1

print(f"Episode completed in {total_steps} steps.\nTotal rewards accumulated: {total_reward}")
```

# **Monitor** 📺

The **Monitor** functionality in Gym shares similarities with wrappers, yet it offers a distinct capability: tracking and recording an agent's performance during interactions with the environment. This tracking can be stored in a file or visualized as a video, providing valuable insights into the agent's behavior and learning progress.

By implementing Gym's **Monitor** feature, users gain the ability to:

1. **Record Performance**: The **Monitor** captures essential information, such as episode rewards, episode lengths, and other relevant metrics, enabling a comprehensive understanding of an agent's performance across episodes.

2. **Logging to File**: It allows the storage of performance metrics in a file, facilitating subsequent analysis, plotting, or data-driven insights into an agent's behavior and learning trends.

3. **Video Generation**: Additionally, the **Monitor** can create video recordings of the agent's interactions within the environment. These videos visually encapsulate the agent's gameplay, offering a tangible representation of its strategies, successes, and failures throughout the learning process.

---
In the newer versions this is deprecated.

```python
from gym.wrappers.monitoring.video_recorder import VideoRecorder

# Initilize the name
env = gym.make(env_id)
env.reset()

video = VideoRecorder(env, "GamePlay.mp4")

total_reward = 0.0
total_steps = 0

done = False
while not done:

    action = env.action_space.sample()
    _, reward, done, _ = env.step(action)
    video.capture_frame()

    total_reward += reward
    total_steps += 1

video.close()

print(f"Episode completed in {total_steps} steps.\nTotal rewards accumulated: {total_reward}")
```
